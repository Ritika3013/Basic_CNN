```python
import numpy as np
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
```

. Plot the image 

```python
def plot_image(img: np.array):
    plt.figure(figsize=(6, 6))
    plt.imshow(img, cmap='gray');
```

```python
def plot_two_images(img1: np.array, img2: np.array):
    _, ax = plt.subplots(1, 2, figsize=(12, 6))
    ax[0].imshow(img1, cmap='gray')
    ax[1].imshow(img2, cmap='gray');
```

```python

img = Image.open('PetImages/Cat/1.jpg')
img = ImageOps.grayscale(img)
img = img.resize(size=(224, 224))
plot_image(img=img)
```

. Task of convolutional layer is to best extract the features 
   various image operations that can be performed 
            . sharpening  .blurring . outlining
             link :  https://setosa.io/ev/image-kernels/

```python
sharpen = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

blur = np.array([
    [0.0625, 0.125, 0.0625],
    [0.125,  0.25,  0.125],
    [0.0625, 0.125, 0.0625]
])
```

```python
def calculate_target_size(img_size: int, kernel_size: int) -> int:
    num_pixels = 0
    
    #  img size = 224, then up to 223
    for i in range(img_size):
        # Add the kernel      
        added = i + kernel_size
        
        # It must be lower than the image size
        if added <= img_size:
            num_pixels += 1
            
    return num_pixels
```

```python
calculate_target_size(img_size=224, kernel_size=3)
```

```python
def convolve(img: np.array, kernel: np.array) -> np.array:
    
    tgt_size = calculate_target_size(
        img_size=img.shape[0],
        kernel_size=kernel.shape[0]
    )
    # size of kernel 
    k = kernel.shape[0]
    
    #Final image formed stor in colv_img
    convolved_img = np.zeros(shape=(tgt_size, tgt_size))
    
    # Iterate over image rows
    for i in range(tgt_size):
        # Iterate over the image cols
        for j in range(tgt_size):
            # Get the current matrix on which conv fun to apply 
            mat = img[i:i+k, j:j+k]
            
            # Apply the convolution - element-wise multiplication and summation of the result
            # Store the result
            convolved_img[i, j] = np.sum(np.multiply(mat, kernel))
            
    return convolved_img
```

```python
img_sharpened = convolve(img=np.array(img), kernel=sharpen)
```

```python
plot_two_images(
    img1=img, 
    img2=img_sharpened
)
```

 . Colors are off values in the matrix don't range between 0 and 255
 . Replacing all negative values with zeros:

```python
def negative_to_zero(img: np.array) -> np.array:
    # Copy the image
    img = img.copy()
    #if values negative make them 0 
    img[img < 0] = 0
    return img
```

```python
plot_two_images(
    img1=img, 
    img2=negative_to_zero(img=img_sharpened)
)
```

```python
img_blurred = convolve(img=np.array(img), kernel=blur)
plot_two_images(
    img1=img, 
    img2=img_blurred
)
```
